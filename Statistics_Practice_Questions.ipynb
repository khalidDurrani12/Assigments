````python
#1. Generate a list of 100 integers containing values between 90 to 130 and store it in the variable `int_list`.
# After generating the list, find the following:

#  (i) Write a Python function to calculate the mean of a given list of numbers.
#      Create a function to find the median of a list of numbers.

#   (ii) Develop a program to compute the mode of a list of integer.

#   (iii) Implement a function to calculate the weighted mean of a list of values and their corresponding weight.

#   (iv) Write a Python function to find the geometric mean of a list of positive number.

#   (v) Create a program to calculate the harmonic mean of a list of value.

#   (vi) Build a function to determine the midrange of a list of numbers (average of the minimum and maximum.

#   (vii) Implement a Python program to find the trimmed mean of a list, excluding a certain percentage of outliers.

import random
from statistics import mode, median
from collections import Counter
import math

# Generate the list of 100 integers between 90 and 130
int_list = [random.randint(90, 130) for _ in range(100)]

# (i) Function to calculate the mean of a list
def calculate_mean(numbers):
    return sum(numbers) / len(numbers)

# (ii) Function to compute the mode of a list
def calculate_mode(numbers):
    return mode(numbers)

# (iii) Function to calculate the weighted mean of a list
def weighted_mean(values, weights):
    return sum(v * w for v, w in zip(values, weights)) / sum(weights)

# (iv) Function to calculate the geometric mean of a list
def geometric_mean(numbers):
    product = math.prod(numbers)
    return product ** (1 / len(numbers))

# (v) Function to calculate the harmonic mean of a list
def harmonic_mean(numbers):
    return len(numbers) / sum(1 / x for x in numbers)

# (vi) Function to calculate the midrange of a list
def midrange(numbers):
    return (max(numbers) + min(numbers)) / 2

# (vii) Function to calculate the trimmed mean of a list
def trimmed_mean(numbers, trim_percentage):
    n = len(numbers)
    trim_count = int(n * trim_percentage)
    sorted_numbers = sorted(numbers)
    trimmed_numbers = sorted_numbers[trim_count : n - trim_count]
    return sum(trimmed_numbers) / len(trimmed_numbers)

# Calculate and display the statistics
mean_val = calculate_mean(int_list)
median_val = median(int_list)
mode_val = calculate_mode(int_list)
weights = [random.uniform(0.5, 2) for _ in range(100)]  # Random weights for demonstration
weighted_mean_val = weighted_mean(int_list, weights)
geometric_mean_val = geometric_mean(int_list)
harmonic_mean_val = harmonic_mean(int_list)
midrange_val = midrange(int_list)
trimmed_mean_val = trimmed_mean(int_list, 0.1)  # Excluding 10% outliers

(mean_val, median_val, mode_val, weighted_mean_val, geometric_mean_val, harmonic_mean_val, midrange_val, trimmed_mean_val)
````

````python
#2.  Generate a list of 500 integers containing values between 200 to 300 and store it in the variable `int_list2`.
# After generating the list, find the following:

#  (i) Compare the given list of visualization for the given data:

#    1. Frequency & Gaussian distribution
#    2. Frequency smoothened KDE plot
#    3. Gaussian distribution & smoothened KDE plot

#  (ii) Write a Python function to calculate the range of a given list of numbers.

#  (iii) Create a program to find the variance and standard deviation of a list of numbers.

#  (iv) Implement a function to compute the interquartile range (IQR) of a list of values.

#  (v) Build a program to calculate the coefficient of variation for a dataset.

#  (vi) Write a Python function to find the mean absolute deviation (MAD) of a list of numbers.

#  (vii) Create a program to calculate the quartile deviation of a list of values.

#  (viii) Implement a function to find the range-based coefficient of dispersion for a dataset.

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate the list of 500 integers between 200 and 300
int_list2 = [random.randint(200, 300) for _ in range(500)]

# (i) Visualizations for the given data
def plot_visualizations(data):
    plt.figure(figsize=(18, 6))

    # Frequency & Gaussian distribution
    plt.subplot(1, 3, 1)
    sns.histplot(data, kde=False, bins=20, color="blue", label="Frequency")
    sns.kdeplot(data, color="red", label="Gaussian Distribution")
    plt.title("Frequency & Gaussian Distribution")
    plt.legend()

    # Frequency smoothened KDE plot
    plt.subplot(1, 3, 2)
    sns.histplot(data, kde=True, bins=20, color="green")
    plt.title("Frequency Smoothened KDE Plot")

    # Gaussian distribution & smoothened KDE plot
    plt.subplot(1, 3, 3)
    sns.kdeplot(data, color="blue", label="Gaussian Distribution")
    sns.kdeplot(data, bw_adjust=0.5, color="orange", label="Smoothened KDE")
    plt.title("Gaussian Distribution & Smoothened KDE Plot")
    plt.legend()

    plt.tight_layout()
    plt.show()

# (ii) Function to calculate the range of a list
def calculate_range(numbers):
    return max(numbers) - min(numbers)

# (iii) Variance and Standard Deviation
def variance_stddev(numbers):
    variance = np.var(numbers, ddof=1)
    stddev = np.sqrt(variance)
    return variance, stddev

# (iv) Interquartile Range (IQR)
def calculate_iqr(numbers):
    q1 = np.percentile(numbers, 25)
    q3 = np.percentile(numbers, 75)
    return q3 - q1

# (v) Coefficient of Variation (CV)
def coefficient_of_variation(numbers):
    mean_val = np.mean(numbers)
    stddev = np.std(numbers, ddof=1)
    return (stddev / mean_val) * 100

# (vi) Mean Absolute Deviation (MAD)
def mean_absolute_deviation(numbers):
    mean_val = np.mean(numbers)
    mad = np.mean([abs(x - mean_val) for x in numbers])
    return mad

# (vii) Quartile Deviation
def quartile_deviation(numbers):
    iqr = calculate_iqr(numbers)
    return iqr / 2

# (viii) Range-based Coefficient of Dispersion
def range_based_dispersion(numbers):
    range_val = calculate_range(numbers)
    mean_val = np.mean(numbers)
    return range_val / mean_val

# Compute values for int_list2
plot_visualizations(int_list2)
range_val = calculate_range(int_list2)
variance, stddev = variance_stddev(int_list2)
iqr_val = calculate_iqr(int_list2)
cv_val = coefficient_of_variation(int_list2)
mad_val = mean_absolute_deviation(int_list2)
quartile_dev = quartile_deviation(int_list2)
range_dispersion = range_based_dispersion(int_list2)

(range_val, variance, stddev, iqr_val, cv_val, mad_val, quartile_dev, range_dispersion)
````

````python
#3. Write a Python class representing a discrete random variable with methods to calculate its expected value and variance.

class DiscreteRandomVariable:
    def __init__(self, outcomes, probabilities):
        """
        Initialize a discrete random variable.

        :param outcomes: A list of possible outcomes.
        :param probabilities: A list of corresponding probabilities.
        """
        if len(outcomes) != len(probabilities):
            raise ValueError("The number of outcomes must match the number of probabilities.")
        if not math.isclose(sum(probabilities), 1):
            raise ValueError("The sum of probabilities must equal 1.")
        self.outcomes = outcomes
        self.probabilities = probabilities

    def expected_value(self):
        """
        Calculate the expected value (mean) of the random variable.

        :return: Expected value.
        """
        return sum(outcome * prob for outcome, prob in zip(self.outcomes, self.probabilities))

    def variance(self):
        """
        Calculate the variance of the random variable.

        :return: Variance.
        """
        mean = self.expected_value()
        return sum(prob * (outcome - mean) ** 2 for outcome, prob in zip(self.outcomes, self.probabilities))


# Example Usage
outcomes = [0, 1, 2, 3, 4]
probabilities = [0.1, 0.2, 0.3, 0.2, 0.2]
random_variable = DiscreteRandomVariable(outcomes, probabilities)
expected_value = random_variable.expected_value()
variance = random_variable.variance()

(expected_value, variance)
````

````python
#5.  Create a Python function to generate random samples from a given probability distribution (e.g.,
# binomial, Poisson) and calculate their mean and variance.

import numpy as np

def generate_samples(distribution, params, num_samples=1000):
    """
    Generate random samples from a given probability distribution and calculate their mean and variance.

    :param distribution: A string specifying the distribution ('binomial', 'poisson', etc.).
    :param params: A dictionary of parameters for the specified distribution.
    :param num_samples: Number of random samples to generate (default is 1000).
    :return: A tuple containing the samples, mean, and variance.
    """
    if distribution == 'binomial':
        n = params.get('n', 1)
        p = params.get('p', 0.5)
        samples = np.random.binomial(n=n, p=p, size=num_samples)
    elif distribution == 'poisson':
        lam = params.get('lambda', 1)
        samples = np.random.poisson(lam=lam, size=num_samples)
    else:
        raise ValueError("Unsupported distribution. Choose 'binomial' or 'poisson'.")

    mean = np.mean(samples)
    variance = np.var(samples, ddof=1)
    return samples, mean, variance

# Example Usage
binomial_params = {'n': 10, 'p': 0.5}
poisson_params = {'lambda': 3}

# Generate samples for binomial distribution
binomial_samples, binomial_mean, binomial_variance = generate_samples('binomial', binomial_params)

# Generate samples for Poisson distribution
poisson_samples, poisson_mean, poisson_variance = generate_samples('poisson', poisson_params)

(binomial_mean, binomial_variance, poisson_mean, poisson_variance)
````

````python
#6. Write a Python script to generate random numbers from a Gaussian (normal) distribution and compute
# the mean, variance, and standard deviation of the samples.

# Generate random numbers from a Gaussian (normal) distribution
def gaussian_distribution_analysis(mean, std_dev, num_samples=1000):
    """
    Generate random numbers from a Gaussian distribution and compute statistics.

    :param mean: The mean of the Gaussian distribution.
    :param std_dev: The standard deviation of the Gaussian distribution.
    :param num_samples: The number of random samples to generate (default is 1000).
    :return: A tuple containing the samples, mean, variance, and standard deviation of the samples.
    """
    samples = np.random.normal(loc=mean, scale=std_dev, size=num_samples)
    sample_mean = np.mean(samples)
    sample_variance = np.var(samples, ddof=1)
    sample_std_dev = np.sqrt(sample_variance)
    return samples, sample_mean, sample_variance, sample_std_dev

# Example Usage
mean = 0  # Mean of the Gaussian distribution
std_dev = 1  # Standard deviation of the Gaussian distribution
num_samples = 1000

samples, sample_mean, sample_variance, sample_std_dev = gaussian_distribution_analysis(mean, std_dev, num_samples)

(sample_mean, sample_variance, sample_std_dev)
````

````python
#8. Write a Python function to calculate the probability density function (PDF) of a continuous random
# variable for a given normal distribution.

from math import exp, sqrt, pi

def normal_pdf(x, mean, std_dev):
    """
    Calculate the probability density function (PDF) of a normal distribution.

    :param x: The value at which to evaluate the PDF.
    :param mean: The mean of the normal distribution.
    :param std_dev: The standard deviation of the normal distribution.
    :return: The value of the PDF at x.
    """
    coefficient = 1 / (std_dev * sqrt(2 * pi))
    exponent = exp(-((x - mean) ** 2) / (2 * std_dev ** 2))
    return coefficient * exponent

# Example Usage
x_value = 0
mean = 0
std_dev = 1

pdf_value = normal_pdf(x_value, mean, std_dev)
pdf_value
````

````python
#9. Create a program to calculate the cumulative distribution function (CDF) of exponential distribution.

from math import exp

def exponential_cdf(x, rate):
    """
    Calculate the cumulative distribution function (CDF) of an exponential distribution.

    :param x: The value at which to evaluate the CDF (x >= 0).
    :param rate: The rate parameter (lambda) of the exponential distribution (rate > 0).
    :return: The value of the CDF at x.
    """
    if x < 0:
        return 0.0
    return 1 - exp(-rate * x)

# Example Usage
x_value = 2
rate = 1.5  # Lambda parameter of the exponential distribution

cdf_value = exponential_cdf(x_value, rate)
cdf_value
````

````python
#10. Write a Python function to calculate the probability mass function (PMF) of Poisson distribution.

from math import exp, factorial

def poisson_pmf(k, lam):
    """
    Calculate the probability mass function (PMF) of a Poisson distribution.

    :param k: The number of occurrences (non-negative integer).
    :param lam: The expected number of occurrences (lambda parameter, lam > 0).
    :return: The probability of observing k occurrences.
    """
    if k < 0 or not isinstance(k, int):
        raise ValueError("k must be a non-negative integer.")
    if lam <= 0:
        raise ValueError("Lambda (lam) must be greater than 0.")
    return (lam ** k) * exp(-lam) / factorial(k)

# Example Usage
k_value = 3
lambda_value = 2.5

pmf_value = poisson_pmf(k_value, lambda_value)
pmf_value
````

````python
#15. A company is testing two different website layouts to see which one leads to higher click-through rates.
# Write a Python function to perform an A/B test analysis, including calculating the t-statistic, degrees of
# freedom, and p-value.

# Use the following data:
# ```python
# layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
# layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

import numpy as np
from scipy import stats

def ab_test(layout_a_clicks, layout_b_clicks):
    # Calculate the sample means
    mean_a = np.mean(layout_a_clicks)
    mean_b = np.mean(layout_b_clicks)

    # Calculate the sample standard deviations
    std_a = np.std(layout_a_clicks, ddof=1)  # ddof=1 for sample standard deviation
    std_b = np.std(layout_b_clicks, ddof=1)

    # Calculate the sample sizes
    n_a = len(layout_a_clicks)
    n_b = len(layout_b_clicks)

    # Calculate the t-statistic
    t_stat = (mean_a - mean_b) / np.sqrt((std_a**2 / n_a) + (std_b**2 / n_b))

    # Calculate the degrees of freedom
    df = ((std_a**2 / n_a + std_b**2 / n_b)**2) / (((std_a**2 / n_a)**2 / (n_a - 1)) + ((std_b**2 / n_b)**2 / (n_b - 1)))

    # Calculate the p-value (two-tailed)
    p_value = stats.t.sf(np.abs(t_stat), df) * 2  # Two-tailed test

    return t_stat, df, p_value

# Data for layouts A and B
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

# Perform the A/B test
t_stat, df, p_value = ab_test(layout_a_clicks, layout_b_clicks)

t_stat, df, p_value
````

````python
#16. A pharmaceutical company wants to determine if a new drug is more effective than an existing drug in
#reducing cholesterol levels. Create a program to analyze the clinical trial data and calculate the tstatistic and p-value for the treatment effect.


#Use the following data of cholestrol level:
#```python
#existing_drug_levels = [180, 182, 175, 185, 178, 176, 172, 184, 179, 183]
#new_drug_levels = [170, 172, 165, 168, 175, 173, 170, 178, 172, 176]```

import numpy as np
from scipy import stats

def analyze_drug_effect(existing_drug_levels, new_drug_levels):
    # Calculate the sample means
    mean_existing = np.mean(existing_drug_levels)
    mean_new = np.mean(new_drug_levels)

    # Calculate the sample standard deviations
    std_existing = np.std(existing_drug_levels, ddof=1)  # ddof=1 for sample standard deviation
    std_new = np.std(new_drug_levels, ddof=1)

    # Calculate the sample sizes
    n_existing = len(existing_drug_levels)
    n_new = len(new_drug_levels)

    # Calculate the t-statistic
    t_stat = (mean_existing - mean_new) / np.sqrt((std_existing**2 / n_existing) + (std_new**2 / n_new))

    # Calculate the degrees of freedom
    df = ((std_existing**2 / n_existing + std_new**2 / n_new)**2) / (((std_existing**2 / n_existing)**2 / (n_existing - 1)) + ((std_new**2 / n_new)**2 / (n_new - 1)))

    # Calculate the p-value (one-tailed test)
    p_value = stats.t.sf(t_stat, df)  # One-tailed test

    return t_stat, df, p_value

# Data: Cholesterol levels for the existing drug and new drug
existing_drug_levels = [180, 182, 175, 185, 178, 176, 172, 184, 179, 183]
new_drug_levels = [170, 172, 165, 168, 175, 173, 170, 178, 172, 176]

# Perform the analysis
t_stat, df, p_value = analyze_drug_effect(existing_drug_levels, new_drug_levels)

t_stat, df, p_value
````

````python
#17.  A school district introduces an educational intervention program to improve math scores. Write a Python
#function to analyze pre- and post-intervention test scores, calculating the t-statistic and p-value to
#determine if the intervention had a significant impact.


#Use the following data of test score:
#  ```python
#  pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
#  post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]
#  ```

import numpy as np
from scipy import stats

def analyze_intervention(pre_scores, post_scores):
    # Calculate the differences between post-intervention and pre-intervention scores
    differences = np.array(post_scores) - np.array(pre_scores)

    # Calculate the mean of the differences
    mean_diff = np.mean(differences)

    # Calculate the standard deviation of the differences
    std_diff = np.std(differences, ddof=1)  # ddof=1 for sample standard deviation

    # Calculate the number of pairs
    n = len(differences)

    # Calculate the t-statistic
    t_stat = mean_diff / (std_diff / np.sqrt(n))

    # Calculate the degrees of freedom
    df = n - 1

    # Calculate the p-value (two-tailed test)
    p_value = stats.t.sf(np.abs(t_stat), df) * 2  # Two-tailed test

    return t_stat, df, p_value

# Data: Pre- and post-intervention test scores
pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

# Perform the analysis
t_stat, df, p_value = analyze_intervention(pre_intervention_scores, post_intervention_scores)

t_stat, df, p_value
````

````python
#18.  An HR department wants to investigate if there's a gender-based salary gap within the company. Develop
#a program to analyze salary data, calculate the t-statistic, and determine if there's a statistically
#significant difference between the average salaries of male and female employees.


#Use the below code to generate synthetic data:
#```python
## Generate synthetic salary data for male and female employees
#np.random.seed(0)  # For reproducibility
#male_salaries = np.random.normal(loc=50000, scale=10000, size=20)
#female_salaries = np.random.normal(loc=55000, scale=9000, size=20)```

import numpy as np
from scipy import stats

def analyze_salary_gap(male_salaries, female_salaries):
    # Calculate the sample means
    mean_male = np.mean(male_salaries)
    mean_female = np.mean(female_salaries)

    # Calculate the sample standard deviations
    std_male = np.std(male_salaries, ddof=1)  # ddof=1 for sample standard deviation
    std_female = np.std(female_salaries, ddof=1)

    # Calculate the sample sizes
    n_male = len(male_salaries)
    n_female = len(female_salaries)

    # Calculate the t-statistic
    t_stat = (mean_male - mean_female) / np.sqrt((std_male**2 / n_male) + (std_female**2 / n_female))

    # Calculate the degrees of freedom
    df = ((std_male**2 / n_male + std_female**2 / n_female)**2) / (((std_male**2 / n_male)**2 / (n_male - 1)) + ((std_female**2 / n_female)**2 / (n_female - 1)))

    # Calculate the p-value (two-tailed test)
    p_value = stats.t.sf(np.abs(t_stat), df) * 2  # Two-tailed test

    return t_stat, df, p_value

# Generate synthetic salary data for male and female employees
np.random.seed(0)  # For reproducibility

male_salaries = np.random.normal(loc=50000, scale=10000, size=20)
female_salaries = np.random.normal(loc=55000, scale=9000, size=20)

# Perform the analysis
t_stat, df, p_value = analyze_salary_gap(male_salaries, female_salaries)

t_stat, df, p_value
````

````python
#21. A political analyst wants to determine if there is a significant association between age groups and voter
# preferences (Candidate A or Candidate B). They collect data from a sample of 500 voters and classify
# them into different age groups and candidate preferences. Perform a Chi-Square test to determine if
# there is a significant association between age groups and voter preferences.


# Use the below code to generate data:
# ```python
# np.random.seed(0)
# age_groups = np.random.choice([ 18 30 , 31 50 , 51+', 51+'], size=30)
# voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)

import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Generate synthetic data
np.random.seed(0)

# Defining the age groups and voter preferences
age_groups = np.random.choice(['18-30', '31-50', '51+'], size=500)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=500)

# Create a contingency table
contingency_table = pd.crosstab(age_groups, voter_preferences)

# Perform Chi-Square test
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

# Output the results
chi2_stat, p_value, dof, expected
````
