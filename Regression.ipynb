**10. Key Assumptions of Multiple Linear Regression**

* **Linearity:** The relationship between
ùëå and predictors is linear.

* **Independence:** Residuals are independent.

* **Homoscedasticity:** Constant variance of residuals.

* **Normality:** Residuals are normally distributed.

* **No Multicollinearity:** Predictors are not highly correlated.

**11. What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model?**

* **Heteroscedasticity** occurs when the variance of residuals (errors) is not constant across all levels of the independent variable(s).

* **Effect on the model:**

  1. Reduces the efficiency of coefficient estimates, leading to less reliable
ùë°-tests and ùëù-values.
  2. Undermines the validity of confidence intervals, making it harder to trust hypothesis tests.

  3. Potentially leads to biased or inefficient regression results.

**12. How can you improve a Multiple Linear Regression model with high multicollinearity?**

* **Multicollinearity** occurs when independent variables are highly correlated.

* **To address it:**

  1. **Remove highly correlated variables:** Use correlation matrices or VIF (Variance Inflation Factor) to identify problematic variables.

  2. **Combine variables:** Create composite or principal components (PCA) to reduce dimensionality.

  3. **Regularization techniques:** Use Ridge Regression or Lasso Regression to shrink coefficients of collinear variables.

  4. **Increase sample size:** Larger datasets can mitigate the impact of multicollinearity.

**13. What are some common techniques for transforming categorical variables for use in regression models?**

* **One-hot encoding:** Converts categories into binary columns (e.g., using pandas get_dummies).

* **Label encoding:** Assigns numerical labels to categories (e.g., 0, 1, 2).

* **Binary encoding:** Combines label encoding with binary conversion for high-cardinality variables.

* **Frequency encoding:** Replaces categories with their frequency of occurrence.

* **Target encoding:** Replaces categories with their mean target value (used cautiously in cross-validation).

**20. How can heteroscedasticity be identified in residual plots, and why is it important to address it?**

* **Identification:**

  1. Residual vs. fitted value plot: Look for a "funnel" shape (widening or narrowing of residuals).

  2. Breusch-Pagan or White‚Äôs test: Statistical tests for heteroscedasticity.

* **Why address it:**

  * Ensures valid hypothesis tests and confidence intervals.
  * Improves efficiency of coefficient estimates.
  * Prevents biased inferences and ensures accurate predictions.

**21. What does it mean if a Multiple Linear Regression model has a high ùëÖ2 but low adjusted ùëÖ2?**

* **High ùëÖ2:** Indicates the model explains a significant proportion of the variance in the dependent variable.

* **Low adjusted ùëÖ2:** Suggests that adding more predictors increases ùëÖ2 without significantly improving the model‚Äôs explanatory power.

* **Implication:** Some predictors may be irrelevant or redundant, leading to overfitting. Adjusted ùëÖ2 penalizes for adding unnecessary variables.

**22. Why is it important to scale variables in Multiple Linear Regression?**

* Scaling ensures variables are on comparable scales, which is crucial because:

  1. **Interpretability:** Unscaled coefficients can be misleading when variables have different units.

  2. **Numerical stability:** Scaled variables prevent large differences in magnitude, reducing computational errors.

  3. **Regularization methods:** Techniques like Ridge or Lasso Regression require scaled data to assign appropriate penalties to coefficients.


**23. What is Polynomial Regression?**

* Polynomial Regression models a non-linear relationship between the dependent and independent variable(s) by transforming predictors into higher-degree polynomials.

* It extends the linear regression framework with terms like ùëã2, ùëã3, etc., while still being a type of regression.

**24. How does Polynomial Regression differ from Linear Regression?

* **Linear Regression:** Assumes a straight-line relationship (ùëå=ùëöùëã+ùëê).

* **Polynomial Regression:** Models more complex, curved relationships using higher-order terms
(e.g., ùëå=ùëè0+ùëè1ùëã+ùëè2ùëã2).


**25. When is Polynomial Regression used?**

* When the relationship between the independent and dependent variable(s) is non-linear.

* **Common applications include:**

  * Fitting non-linear trends in data.

  * Modeling growth rates, natural processes, or phenomena with quadratic/cubic behavior.

**28. What are the limitations of Polynomial Regression?**

* **1.Overfitting:** Higher-degree polynomials may fit the training data too closely, reducing generalization.

* **2.Complexity:** Harder to interpret compared to linear regression.

* **3.Extrapolation:** Predictions beyond the range of data can behave erratically.

* **4.Collinearity:** Higher-order terms can lead to multicollinearity.

* **5.Computational cost:** Increases with the degree of the polynomial and number of predictors.


**29. What methods can be used to evaluate model fit when selecting the degree of a polynomial?**

**1. Cross-validation:** Assess performance on unseen data to prevent overfitting.

**2. Adjusted ùëÖ2:** Accounts for the number of predictors.

**3. AIC/BIC (Akaike/Bayesian Information Criterion):** Penalize model complexity to balance fit and simplicity.

**4. Residual analysis:** Inspect residual plots to check for randomness.

**5. Visualization:** Compare fitted curves to actual data.
