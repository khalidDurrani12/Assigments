```python
# 1. Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a threedimensional space.
# np.random.seed(30)
#data = {
#    'X': np.random.uniform(-10, 10, 300),
#    'Y': np.random.uniform(-10, 10, 300),
#   'Z': np.random.uniform(-10, 10, 300)
#}
#df = pd.DataFrame(data)

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd

# Generate the dataset (as given)
np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}

# Convert to a DataFrame
df = pd.DataFrame(data)

# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the points
ax.scatter(df['X'], df['Y'], df['Z'], c='blue', marker='o')

# Add labels and title
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Scatter Plot of Random Data Points')

# Show plot
plt.show()
```

```python
# 2.  Using the Student Grades, create a violin plot to display the distribution of scores across different grade
#categories.np.random.seed(15)
#data = {
#    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
#    'Score': np.random.randint(50, 100, 200)
#}
#df = pd.DataFrame(data)

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate Student Grades dataset
np.random.seed(15)
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}

df_grades = pd.DataFrame(data)

# Create a violin plot to display the distribution of scores across grade categories
plt.figure(figsize=(8, 6))
sns.violinplot(x='Grade', y='Score', hue='Grade', data=df_grades, palette='Set2')

# Add title and labels
plt.title('Distribution of Scores Across Grades')
plt.xlabel('Grade')
plt.ylabel('Score')

# Show plot
plt.grid(True)
plt.show()
```

```python
# 3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.
#np.random.seed(20)
#data = {
#    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
#    'Day': np.random.choice(range(1, 31), 100),
#    'Sales': np.random.randint(1000, 5000, 100)
#}
#df = pd.DataFrame(data)

# Generate Sales dataset
np.random.seed(20)
data_sales = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

df_sales = pd.DataFrame(data_sales)

# Pivot the data to create a matrix for heatmap
sales_pivot = df_sales.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

# Create a heatmap to visualize sales across different months and days
plt.figure(figsize=(10, 8))
sns.heatmap(sales_pivot, annot=True, cmap='Blues', fmt=".1f")

# Add title and labels
plt.title('Sales Variation Across Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')

# Show plot
plt.show()
```

```python
# 4. Using the given x and y data, generate a 3D surface plot to visualize the function .

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D surface plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(df['X'].values.reshape(100, 100),
                df['Y'].values.reshape(100, 100),
                df['Z'].values.reshape(100, 100),
                cmap='viridis', edgecolor='none')

# Add labels and title
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Surface Plot of z = sin(sqrt(x^2 + y^2))')

# Show plot
plt.show()
```

```python
# 5. Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (xaxis), and bubble size proportional to the populationnp.random.seed(25)
# data = {
#    'Country': ['USA', 'Canada', 'UK',
#    'Germany', 'France'],
#    'Population':np.random.randint(100, 1000, 5),
#    'GDP': np.random.randint(500, 2000,5)
# }
# df = pd.DataFrame(data)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Seed for reproducibility
np.random.seed(25)

# Create dataset
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}

df = pd.DataFrame(data)

# Create a bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(df['GDP'], df['Population'], s=df['Population']*10, alpha=0.5, c='blue', edgecolors='w', linewidth=2)

# Add titles and labels
plt.title('Bubble Chart of Country Population vs. GDP')
plt.xlabel('GDP (in billions)')
plt.ylabel('Population (in millions)')

# Annotate the bubbles with country names
for i, country in enumerate(df['Country']):
    plt.annotate(country, (df['GDP'][i], df['Population'][i]), fontsize=10, ha='center')

plt.grid(True)
plt.show()
```
