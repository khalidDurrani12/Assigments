```python
# 1. Create a scatter plot to visualize the relationship between two variables, by generating a synthetic dataset.

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate synthetic dataset
np.random.seed(42)  # For reproducibility
x = np.random.rand(100) * 10  # 100 random points between 0 and 10
y = 2.5 * x + np.random.randn(100) * 5  # Linear relation with some noise

# Create a scatter plot using Seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(x=x, y=y, color='purple', marker='o')

# Add title and labels
plt.title('Scatter Plot of Synthetic Data')
plt.xlabel('x values')
plt.ylabel('y values')

# Show plot
plt.grid(True)
plt.show()
```

```python
# 2. Generate a dataset of random numbers. Visualize the distribution of a numerical variable.

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate synthetic dataset of random numbers from a normal distribution
np.random.seed(42)  # For reproducibility
data = np.random.randn(1000)  # 1000 random numbers from a standard normal distribution

# Create a histogram with Seaborn to visualize the distribution
plt.figure(figsize=(8, 6))
sns.histplot(data, bins=30, kde=True, color='purple')

# Add title and labels
plt.title('Distribution of Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show plot
plt.grid(True)
plt.show()
```

```python
# 3. Create a dataset representing categories and their corresponding values. Compare different categories based on numerical values.

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Create dataset
data = {
    'Category': ['Category A', 'Category B', 'Category C', 'Category D', 'Category E'],
    'Value': [25, 40, 15, 30, 20]
}

# Convert the dataset into a DataFrame
df = pd.DataFrame(data)

# Create a bar plot to compare categories based on numerical values
plt.figure(figsize=(8, 6))
sns.barplot(x='Category', y='Value',hue='Category', data=df, palette='coolwarm', legend = False)

# Add title and labels
plt.title('Comparison of Categories Based on Values')
plt.xlabel('Category')
plt.ylabel('Value')

# Show plot
plt.grid(True)
plt.show()
```

```python
# 4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Generate dataset
np.random.seed(42)  # For reproducibility
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [np.random.normal(loc, 10, 100) for loc in [20, 35, 50, 65]]  # Different normal distributions for each category

# Convert data to a DataFrame
df = pd.DataFrame({
    'Category': np.repeat(categories, 100),
    'Value': np.concatenate(values)
})

# Create a violin plot to visualize the distribution of values across categories
plt.figure(figsize=(10, 6))
sns.violinplot(x='Category', y='Value', hue = 'Category', data=df, palette='Set2')

# Add title and labels
plt.title('Distribution of Values Across Categories (Violin Plot)')
plt.xlabel('Category')
plt.ylabel('Value')

# Show plot
plt.grid(True)
plt.show()
```

```python
# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Generate synthetic dataset with correlated features
np.random.seed(42)
n_samples = 100

# Generate random features
X1 = np.random.normal(10, 2, n_samples)  # Feature 1
X2 = 2 * X1 + np.random.normal(0, 2, n_samples)  # Correlated Feature 2
X3 = 0.5 * X1 + np.random.normal(0, 1, n_samples)  # Correlated Feature 3
X4 = np.random.normal(5, 1, n_samples)  # Independent Feature

# Combine into a DataFrame
df = pd.DataFrame({
    'Feature 1': X1,
    'Feature 2': X2,
    'Feature 3': X3,
    'Feature 4': X4
})

# Compute the correlation matrix
corr_matrix = df.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5, vmin=-1, vmax=1)

# Add title
plt.title('Correlation Matrix Heatmap')

# Show plot
plt.show()
```
