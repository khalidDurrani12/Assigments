**1. Difference Between numpy.array(), numpy.asarray(), and numpy.asanyarray()**

**a. ``numpy.array()``**

**Purpose:** Creates a new NumPy array from any input, copying the data unless explicitly told not to (copy=False).

**Behavior:** Even if the input is already a NumPy array, it can still create a new copy of the array, depending on the copy argument.

**Usage:**

    import numpy as np
    lst = [1, 2, 3]
    arr = np.array(lst)
    print(arr)  # Creates a new array from the list

**b.``numpy.asarray()``**

**Purpose:** Converts input into a NumPy array without copying data if the input is already a NumPy array.

**Behavior:** If the input is a NumPy array, it returns the original array without making a copy (unless dtype is different).

**Usage:**

    lst = [1, 2, 3]
    arr = np.asarray(lst)
    print(arr)  # Creates an array without copying if possible

**Key Point:** If the input is already a NumPy array with the same dtype, it does not create a new copy, saving memory.

**c.``numpy.asanyarray()``**

**Purpose:** Similar to ``asarray()``, but it returns the input as an array or a subclass of an array (e.g., matrix, masked_array).

**Behavior:** If the input is a subclass of ndarray (like matrix),
``asanyarray()`` will return the subclass type, while ``asarray()`` would convert it to a base ndarray.

**Usage:**

    matrix = np.matrix([[1, 2], [3, 4]])
    arr = np.asanyarray(matrix)
    print(type(arr))  # Returns numpy.matrix (subclass)

**2. Difference Between Deep Copy and Shallow Copy**

**a. Shallow Copy:**

**Definition:** A shallow copy creates a new object, but does not copy the nested objects (e.g., elements of a list or an array). Instead, it copies references to the objects, meaning changes in the nested objects will be reflected in both the original and copied objects.

**Behavior:** The outer object is copied, but the inner objects (references) are shared between the original and copied objects.

**Usage Example:**

    import copy
    list1 = [[1, 2], [3, 4]]
    shallow_copy = copy.copy(list1)
    shallow_copy[0][0] = 10  # Modifies both shallow_copy and list1
    print(list1)  # Output: [[10, 2], [3, 4]]

**b. Deep Copy:**

**Definition:** A deep copy creates a new object and recursively copies all nested objects, meaning changes to nested objects in the copied version do not affect the original object.

**Behavior:** Both the outer and inner objects are fully copied, and modifications in the copy do not affect the original object.

**Usage Example:**

    deep_copy = copy.deepcopy(list1)
    deep_copy[0][0] = 99  # Only modifies deep_copy, not list1
    print(list1)  # Output: [[1, 2], [3, 4]]

```python
# 14. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and
# Perform the following tasks using Matplotlib:

#a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').
# b) Label the x-axis as 'Date'.
# c) Set the title of the plot as 'Temperature and Humidity Over Time'.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Step 1: Create a date range
date_range = pd.date_range(start='2023-01-01', periods=100, freq='D')

# Step 2: Generate random temperature and humidity data
np.random.seed(0)  # For reproducibility
temperature = np.random.uniform(low=15, high=30, size=len(date_range))  # Temperature between 15 and 30 degrees
humidity = np.random.uniform(low=30, high=90, size=len(date_range))      # Humidity between 30% and 90%

# Step 3: Create a DataFrame
df = pd.DataFrame({
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity
})

# Step 4: Plotting
fig, ax1 = plt.subplots(figsize=(12, 6))  # Increase the figure size for better clarity

# Plot Temperature
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='tab:red')
ax1.plot(df['Date'], df['Temperature'], color='tab:red', label='Temperature')
ax1.tick_params(axis='y', labelcolor='tab:red')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.set_ylabel('Humidity (%)', color='tab:blue')
ax2.plot(df['Date'], df['Humidity'], color='tab:blue', label='Humidity')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Format x-axis for better readability
ax1.xaxis.set_major_locator(mdates.WeekdayLocator(interval=1))  # Show one tick per week
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))
```

```python
# 15. Create a NumPy array data containing 1000 samples from a normal distribution.
# Perform the following tasks using Matplotlib:
# a) Plot a histogram of the data with 30 bins.

# b) Overlay a line plot representing the normal distribution's probability density function (PDF).

# c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

# d) Set the title of the plot as 'Histogram with PDF Overlay'.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Create a NumPy array with 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Step 2: Plot a histogram of the data with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# Step 3: Overlay the PDF of the normal distribution
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = (1 / (np.sqrt(2 * np.pi))) * np.exp(-0.5 * x**2)
plt.plot(x, p, 'k', linewidth=2, label='Normal PDF')

# Step 4: Label the axes and set the title
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')
plt.legend()

# Show the plot
plt.show()
```

```python
# 16. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
# origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Create two random arrays
np.random.seed(0)  # For reproducibility
x = np.random.rand(100) * 10 - 5  # Random values between -5 and 5
y = np.random.rand(100) * 10 - 5  # Random values between -5 and 5

# Step 2: Determine the color based on quadrant
colors = []
for i in range(len(x)):
    if x[i] >= 0 and y[i] >= 0:
        colors.append('Quadrant I')
    elif x[i] < 0 and y[i] >= 0:
        colors.append('Quadrant II')
    elif x[i] < 0 and y[i] < 0:
        colors.append('Quadrant III')
    else:
        colors.append('Quadrant IV')

# Step 3: Create a scatter plot
sns.scatterplot(x=x, y=y, hue=colors, palette='Set1')

# Step 4: Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')
plt.legend(title='Quadrants')

# Show the plot
plt.show()
```

```python
# 18. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
# values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'8

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
import numpy as np

# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 100, size=len(categories))

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(Category=categories, Value=values))

# Create a figure
p = figure(x_range=categories, title='Random Categorical Bar Chart',
           toolbar_location=None, tools="")

# Add bars
p.vbar(x='Category', top='Value', source=source, color='navy', width=0.9)

# Add hover tool
hover_tool = HoverTool()
hover_tool.tooltips = [("Category", "@Category"), ("Value", "@Value")]
p.add_tools(hover_tool)

# Label axes
p.xaxis.axis_label = "Categories"
p.yaxis.axis_label = "Values"
p.xgrid.grid_line_color = None
p.y_range.start = 0

# Show the plot
show(p)
```
